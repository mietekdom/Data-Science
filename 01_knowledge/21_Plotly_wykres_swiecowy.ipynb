```
import pandas as pd
import plotly
import plotly.graph_objects as go
```

```
def fetch_financial_data(company='GE'):
  import pandas_datareader.data as web
  return web.DataReader(name=company, data_source='stooq')
  

df = fetch_financial_data()
df = df.reset_index()
df = df[df.Date > '2020-01-01']
df.head(5)
```

```
fig = go.Figure(data=go.Candlestick(x=df.Date, open=df.Open, high=df.High, low=df.Low, close=df.Close),layout=go.Layout(title_text='General Electric stock market in 2020'))
fig.show()
```
